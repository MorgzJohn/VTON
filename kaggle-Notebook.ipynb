## Step 3: Fix PositionNet Import Error
IDM-VTON references PositionNet which was removed in newer diffusers versions.
This patches the source file to remove the broken reference.

```python
with open('src/unet_hacked_garmnet.py', 'r') as f:
    content = f.read()
content = content.replace(', PositionNet', '')
content = content.replace('PositionNet,', '')
content = content.replace('PositionNet', '')
content = content.replace(
    'self.position_net = (\n                positive_len=positive_len, out_dim=cross_attention_dim, feature_type=feature_type\n            )',
    'self.position_net = None'
)
with open('src/unet_hacked_garmnet.py', 'w') as f:
    f.write(content)
print("PositionNet fixed!")
```

## Step 4: Fix Gradio Client Bug
Patches a TypeError in gradio_client/utils.py where schema is sometimes
a bool instead of a dict, causing internal server errors in the interface.

```python
with open('/usr/local/lib/python3.12/dist-packages/gradio_client/utils.py', 'r') as f:
    lines = f.readlines()

# Find and fix ALL .get() calls and "in schema" checks in the file
new_lines = []
for i, line in enumerate(lines):
    # Fix any line with schema.get( that doesn't have isinstance check
    if 'schema.get(' in line and 'isinstance' not in line:
        indent = len(line) - len(line.lstrip())
        spaces = ' ' * indent
        new_lines.append(line.replace('schema.get(', '(schema if isinstance(schema, dict) else {}).get('))
    # Fix any line with "in schema" that doesn't have isinstance check  
    elif 'in schema' in line and 'isinstance' not in line and 'not in schema' not in line:
        new_lines.append(line.replace('in schema', 'in (schema if isinstance(schema, dict) else {})'))
    else:
        new_lines.append(line)

with open('/usr/local/lib/python3.12/dist-packages/gradio_client/utils.py', 'w') as f:
    f.writelines(new_lines)

print("Done! All schema references patched.")
```

## Step 6: Add Memory Optimizations to app.py
Adds attention slicing, VAE slicing and tiling to reduce VRAM usage.
Required because T4 GPU has 15GB but model needs ~14.5GB.

```python
with open('/kaggle/working/IDM-VTON/gradio_demo/app.py', 'r') as f:
    content = f.read()

# Add memory optimizations after pipe is loaded
content = content.replace(
    'pipe.to(device)',
    '''pipe.to(device)
pipe.enable_attention_slicing(1)
pipe.enable_vae_slicing()
pipe.enable_vae_tiling()
torch.cuda.empty_cache()'''
)

with open('/kaggle/working/IDM-VTON/gradio_demo/app.py', 'w') as f:
    f.write(content)
print("Done!")
```

```python
with open('/kaggle/working/IDM-VTON/gradio_demo/app.py', 'r') as f:
    content = f.read()

# Fix the bad indentation
content = content.replace(
    '    pipe.to(device)\npipe.enable_attention_slicing(1)\npipe.enable_vae_slicing()\npipe.enable_vae_tiling()\ntorch.cuda.empty_cache()\n',
    '    pipe.to(device)\n    pipe.enable_attention_slicing(1)\n    pipe.enable_vae_slicing()\n    pipe.enable_vae_tiling()\n    torch.cuda.empty_cache()\n'
)

with open('/kaggle/working/IDM-VTON/gradio_demo/app.py', 'w') as f:
    f.write(content)
print("Fixed!")
```

## Step 9: Add Cache Clearing Before VAE Decode
Inserts torch.cuda.empty_cache() before the VAE decode step in
tryon_pipeline.py to free reserved but unallocated GPU memory.

```python
for i, line in enumerate(lines):
    if 'vae.decode(latents' in line:
        indent = len(line) - len(line.lstrip())
        spaces = ' ' * indent
        lines[i] = spaces + 'torch.cuda.empty_cache()\n' + line
        print(f"Fixed line {i+1}")

with open('/kaggle/working/IDM-VTON/src/tryon_pipeline.py', 'w') as f:
    f.writelines(lines)
print("Done!")
```

## Step 10: Offload UNet to CPU Before Decode
Moves UNet and UNet encoder to CPU before VAE decode to free ~8GB VRAM.
This is the critical fix that allows the final image generation to complete.

```python
with open('/kaggle/working/IDM-VTON/src/tryon_pipeline.py', 'r') as f:
    lines = f.readlines()

for i, line in enumerate(lines):
    if i+1 == 1876:
        lines[i] = '''            # Offload UNet to free VRAM for decode
            self.unet.to('cpu')
            self.unet_encoder.to('cpu')
            torch.cuda.empty_cache()
            import gc
            gc.collect()
            torch.cuda.synchronize()
'''
        print("Fixed!")
        break

with open('/kaggle/working/IDM-VTON/src/tryon_pipeline.py', 'w') as f:
    f.writelines(lines)
```

## Step 11: Set CUDA Memory Configuration
Configures PyTorch memory allocator to use expandable segments,
